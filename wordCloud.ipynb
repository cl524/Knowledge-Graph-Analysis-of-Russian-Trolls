```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd, string

from PIL import Image, ImageFont, ImageDraw
```

```python
data_file3='final_result_1_p-Copy_stop_eli_3cluster.csv'
df3=pd.read_csv( data_file3, header=0, encoding="ISO-8859â€“1", engine='python' )
```

```python
questions0=df3[df3['Group']==0]['text_P'].tolist()

ssin0=""

i=0

for ssi in questions0:
    ssin0 = ssin0 + str(ssi)
    ssin0 += " "

ssin0.strip()

wordcloud0 = WordCloud(width = 6000, height = 4000, background_color='white', max_words=1000000, collocations=False).generate(ssin0)

plt.imshow(wordcloud0, interpolation='bilinear')
plt.show()
```
