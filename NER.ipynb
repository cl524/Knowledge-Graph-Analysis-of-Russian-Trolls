```python
import spacy
import numpy as np
from spacy import displacy
from spacy.symbols import ORTH

from collections import Counter

import en_core_web_sm
nlp = en_core_web_sm.load()

import pandas as pd
import re, csv, string
```

```python
data_file3='final_result_1_p-Copy.csv'
df3=pd.read_csv( data_file3, header=0, encoding="ISO-8859â€“1", engine='python' )

questions3=df3['text_P'].tolist()
```

```python
import string, re

wordCount=0
llist=[]

for ll in questions3:
    senten=str(ll)

    senten = re.sub(r"[{}]".format(string.punctuation), "", senten)
    llist=senten.split(" ")

    wordCount+=len(llist)
```

```python
items3=[]

entity_dic={'PERSON':0, 'NORP':0, 'FAC':0, 'ORG':0, 'GPE':0, 'LOC':0, 'PRODUCT':0, 'EVENT':0, 'WORK_OF_ART':0, 'LAW':0,
           'LANGUAGE':0, 'DATE':0, 'TIME':0, 'PERCENT':0, 'MONEY':0, 'QUANTITY':0, 'ORDINAL':0, 'CARDINAL':0}

print(entity_dic['PERSON'])

for st3 in questions3:
    st3=str(st3)
    st3=re.sub("\'\S+", '', st3)
    
    doc3=nlp((st3))

    items3 = items3 + ( [x3.text for x3 in doc3.ents if x3.label_=='PERSON'] )
    
    for x3 in doc3.ents:
        entity_dic[x3.label_]+=1
```

```python
print(entity_dic)
```

```python
with open('Person_Counting.csv', mode='w', encoding="utf-8") as employee_file:
    fieldnames = ['Person', 'Counting']
    
    employee_writer = csv.DictWriter(employee_file, fieldnames=fieldnames, lineterminator = '\n')
    employee_writer.writeheader()
    employee_writer = csv.writer(employee_file, quotechar='"', quoting=csv.QUOTE_MINIMAL, lineterminator = '\n')
    
    i=0
    while i<len(Counter(items3).most_common()):

        person=(Counter(items3).most_common())[i][0]
        counting=(Counter(items3).most_common())[i][1]
        
        employee_writer.writerow([person, counting])
        i+=1
```
